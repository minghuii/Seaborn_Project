```python

# Pandas for managing datasets
import pandas as pd

# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline

# Seaborn for plotting and styling
import seaborn as sns
```

```python
# Read dataset
df = pd.read_csv('try.csv')
df.head()
```

```python
# a scatter plot
sns.lmplot(x='Likes', y='RTs', data=df)
# Tweak using Matplotlib
```

We can see an outlier here, which with few likes but has huge retweets. That one is " @XHSports: '@pyeongchang2018 have closed with a performance by popular boy band EXO https://t.co/fE6UqYkovu
Number of retweets: 10267"

```python
# a scatter plot
sns.lmplot(x='Likes', y='RTs', data=df)
# Tweak using Matplotlib
plt.ylim(-10, 150)
plt.xlim(-20, 150)
```

From this picture, we can see that genrally more likeswould come with more retweets. 

```python
# Scatterplot arguments
sns.lmplot(x='Likes', y='RTs', data=df,
           fit_reg=False, # No regression line
           hue='Source')   # Color by evolution stage
plt.ylim(-10, 60)
plt.xlim(-20, 150)
```

```python
# Swarm plot with Pokemon color palette
plt.figure(figsize=(12,12))
sns.swarmplot(x='Source', y='len', data=df)
plt.ylim(40, 150)
```

From this picture, we can see that compared with the tweetsfrom Xinhua or Hongkong, Taiwan's tweets are shorter. 

```python
ax=sns.countplot(x="SA", hue="Source", data=df)
total=200
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}'.format(height/total),
            ha="center") 
```
